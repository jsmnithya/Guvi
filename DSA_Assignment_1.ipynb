```python

def reg(username,password):
    a=username.split("@")
    if '.' in a[1] and a[1][0]!=0:
        sc=['~','!','@','#','$','%','^','&','*','(',')','-','_','+','.','/','|',':',';','<','>']
        if a[0][0].isnumeric()==False and a[0][0] not in sc:
            v=validpass(password)
            if v==True:
                f=open('email.txt','a')
                f.write(username)
                f.write(" ")
                f.write(password)
                f.write("\n")
                f.close()
                return True
            else:
                print("Invalid password")
                return False
        else:
            print("Invalid username")
            return False
    else:
        print('Invalid username')
        return False
def validpass(password):
    lp=len(password)
    sc=['~','!','@','#','$','%','^','&','*','(',')','-','_','+','.','/','|',':',';','<','>']
    if lp>5 and lp<16:
        csc=0
        cd=0
        cu=0
        cl=0
        for i in password:
            if i in sc:
                csc=csc+1
            elif i.isnumeric()==True:
                cd=cd+1
            elif i.isupper()==True:
                cu=cu+1
            elif i.islower()==True:
                cl=cl+1
        if csc>0 and cd>0 and cu>0 and cl>0:
                return True
    return False
def login(username,password):
    f=open('email.txt','r')
    for i in f.readlines():
        detail=i.split()
        if username == detail[0]:
            if password==detail[1]:
                print("Welcome")
                f.close()
                return True
    print("Invalid username or password")
    f.close()
    return False

def forget(username):
    f=open('email.txt','r+')
    for i in f.readlines():
        detail=i.split()
        if username == detail[0]:
            print("Enter 1 to retrieve password")
            print("Enter 2 to change the password")
            c=input()
            if c=='1':
                print(detail[1])
                return True
            elif c=='2':
                n=input("Enter the new password:")
                v=validpass(n)
                if v==True:
                    data=f.read()
                    data=data.replace(detail[1],n)
                    f.write(data)
                    print("password changed")
                    f.close()
                    return True
                else:
                    print("invalid password")
                    f.close()
                    return False
            else:
                print("Invalid option")
                f.close()
                return False
    print("Invalid username")
    return False

print("----MENU----")
print("Enter 1 for registration")
print('Enter 2 for login')
a=input()
u = input("Enter the username:")
p = input("Enter the password:")
if a=='1':
    b=reg(u,p)
    if b==True:
        print("successfully registered")
    else:
        print("Registration unsuccessful")
elif a=='2':
    c=login(u, p)
    if c==True:
        print("successfully logged in")
    else:
        print("login unsuccessful")
        print("Do you want to register new or forgat password?")
        print("Enter 1 for registration")
        print("Enter 2 for forget password")
        print("Enter any other value to exit")
        d=input()
        if d=='1':
            u = input("Enter the username:")
            p = input("Enter the password:")
            b=reg(u,p)
            if b==True:
                print("successfully registered")
            else:
                print("Registration unsuccessful")
        elif d=='2':
            username = input("Enter the username:")
            e=forget(username)
```
